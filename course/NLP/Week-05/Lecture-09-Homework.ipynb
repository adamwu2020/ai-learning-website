## Exercise 9.2

Implement document similarity using document embeddings (average of word vectors). Compare with TF-IDF cosine similarity.

```python
# Exercise 9.2 starter: document similarity with word vectors
def doc2vec_avg(text, model):
    """Compute document vector as average of word vectors."""
    tokens = word_tokenize(text.lower())
    vectors = [model.wv[w] for w in tokens if w in model.wv]
    if not vectors:
        return np.zeros(model.vector_size)
    return np.mean(vectors, axis=0)

doc_texts = [
    'Emma loves Mr. Knightley and they marry at the end.',
    'Mr. Woodhouse is the loving father of Emma.',
    'Frank Churchill arrives from Yorkshire with secrets.',
]
# YOUR CODE: compute doc vectors, show similarity matrix,
# then compare with TF-IDF cosine similarity
```
