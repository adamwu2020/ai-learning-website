## Exercise 15.4

Build a simple FAQ chatbot for your institution's website using the `RetrievalChatbot` framework.

```python
# Exercise 15.4
class InstitutionFAQBot:
    """FAQ chatbot using TF-IDF cosine similarity."""

    # YOUR CODE: populate with at least 15 Q&A pairs for your institution
    FAQ = [
        ('What are the application deadlines?', 'Applications are due March 1 for fall admission.'),
        ('How do I reset my student portal password?', 'Visit the IT helpdesk at help.university.edu or call ext. 5555.'),
        # Add more Q&A pairs here...
    ]

    def __init__(self):
        questions, self.answers = zip(*self.FAQ)
        self.vec = TfidfVectorizer(ngram_range=(1, 2))
        self.faq_vecs = self.vec.fit_transform(questions)

    def respond(self, query, threshold=0.15):
        qv = self.vec.transform([query])
        sims = cosine_similarity(qv, self.faq_vecs).flatten()
        idx = sims.argmax()
        if sims[idx] >= threshold:
            return self.answers[idx], sims[idx]
        return 'I could not find an answer. Please contact our support team.', 0.0

faq_bot = InstitutionFAQBot()
queries = [
    'When is the application due?',
    'I forgot my password',
    'What is the tuition cost?',
]
for q in queries:
    ans, score = faq_bot.respond(q)
    print(f'Q: {q}')
    print(f'A: {ans}  (sim={score:.3f})')
    print()
```
