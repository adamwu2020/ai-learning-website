## Exercise 6.1

Write `pos_frequency_analysis(text)` that returns a FreqDist of POS tags. What are the most common tags in a news article?

```python
# Exercise 6.1 starter
def pos_frequency_analysis(text):
    tokens = word_tokenize(text)
    tagged = pos_tag(tokens)
    return nltk.FreqDist(tag for _, tag in tagged)

news_sample = """
The Federal Reserve raised interest rates by 0.25 percent on Wednesday.
The decision was unanimous among board members. Markets fell sharply.
"""
pos_dist = pos_frequency_analysis(news_sample)
print('Top POS tags:')
pos_dist.tabulate()
```
