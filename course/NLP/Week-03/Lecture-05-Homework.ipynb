## Exercise 5.1

Train a Naive Bayes classifier on NLTK `names` corpus (male/female). Use last 2 chars as feature. What accuracy do you get?

```python
# Exercise 5.1 starter
from nltk.corpus import names

def gender_features(name):
    return {
        'last_letter': name[-1].lower(),
        'last_two': name[-2:].lower(),
        'first_letter': name[0].lower(),
        # YOUR CODE: add more features
    }

labeled = [(n, 'male') for n in names.words('male.txt')] + \
          [(n, 'female') for n in names.words('female.txt')]
random.shuffle(labeled)
featuresets = [(gender_features(n), g) for n, g in labeled]
train, test = featuresets[500:], featuresets[:500]
clf = nltk.NaiveBayesClassifier.train(train)
print(f'Gender accuracy: {nltk.classify.accuracy(clf, test):.4f}')
```
